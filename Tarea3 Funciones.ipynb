##Parámetros de las Funciones en Python
Una función puede definir, opcionalmente, una secuencia de parámetros con los que invocarla.

Veamos los conceptos de programación **paso por Valor:** Un lenguaje que utiliza paso por valor de los argumentos, lo que realmente hace es copiar el valor de las variables en los respectivos parámetros. Cualquier modificación del parámetro, no afecta a la variable externa correspondiente.

Programación **Paso por Referencia:** Utiliza paso por referencia, lo que realmente hace es copiar en los parámetros la dirección de memoria de las variables que se usan como argumento. Esto implica que realmente hagan referencia al mismo objeto/elemento y cualquier modificación del valor en el parámetro afectará a la variable externa correspondiente.

En Python todo es un objeto, entonces,  ¿cómo se pasan los argumentos en Python, por valor o por referencia? Lo que ocurre en Python realmente es que se pasa por valor la referencia del objeto. 
¿Qué implicaciones tiene esto? Básicamente que si el tipo que se pasa como argumento es inmutable, cualquier modificación en el valor del parámetro no afectará a la variable externa pero, si es mutable (como una lista o diccionario), sí se verá afectado por las modificaciones.

#Parámetros Posicionales y parámetros con nombre en una función
Recordando que cuando invocamos una función con diferentes argumentos, los valores se asignan a los parámetros en el mismo orden en que se indican.

por ejemplo si llamamos a la función de la siguiente manera:
def (Nombre, Mensaje):
Luego llamamos a la funcion saludo
saludo("Jaime", "Buenos Días"): 
De este modo, al parámetro nombre se le asignará el valor "Jaime" y al parámetro "mensaje" el valor "Buenos Dias"

Esto es así porque el valor en el que se asignan los argumentos depende del orden con el que se llaman. Sin embargo, el orden se puede cambiar si llamamos a la función indicando el nombre de los parámetros. por ejemplo:


```python
def saludo(nombre="Jaime Mercar", mensaje="Buenos Días"):
    print(f'(nombre,mensaje)')
    print(f'mensaje, nombre')
```

```python
saludo(mensaje="Buenos Días", nombre="Jaime")
```

Como vemos, podemos mezclar el orden de los parámetros si indicamos su nombre. Eso sí, siempre deben estar a la derecha de los parámetros posicionales, es decir, aquellos que se indican sin nombre y cuyo valor se asigna en el orden en el que se indica.

1-Por defecto, al llamar a una función los valores de los argumentos se asignan en el mismo orden en el que se pasan al invocar a dicha función.

Se puede modificar el orden de los argumentos con el que se invoca a una función si se indica el nombre de los parámetros. Los parámetros con nombre siempre aparecen después de los posicionales.

#Retorno de multiples valores de una función en python
Recordando que una función sirve para devolver un valor. Este valor puede ser un número, una cadena o cualquier otro tipo de datos.
Acontinuacion veremos los  métodos para crear una función que devuelva múltiples valores. Devolveremos diferentes objetos de datos como una lista, un diccionario y otros objetos de una función definida.



##Uso de **Tupla** para devolver multiples valores de una Función
i devolvemos valores de una función separados por una coma, se consideran una tupla. Las tuplas suelen ir entre paréntesis. En el siguiente código, devolveremos una tupla de una función: 

```python
def Multiples_Valores(a):
    b = a+2
    c = a+4
    return b, c

x = Multiples_Valores(5)
print(x,type(x))
```

##Uso de **Lista** para devolver multiples valores de una Función
Las listas se utilizan para almacenar diferentes elementos con un nombre común y en posiciones específicas. Las funciones también pueden devolver varios valores en una lista, como se muestra a continuación.


```python
def Valores_lista(a):
  X = a**2 + 5
  Y = a + 2
  return [X,Y]
z = Valores_lista (2)
print (z,type(z))
```

##Uso de un **Diccionario** para devolver multiples valores de una Función
En python se utilizan para almacenar pares: Clave_Valor.Podemos tener la salida final en un formato mucho más organizado devolviendo un diccionario de una función con teclas asignadas a diferentes valores.  

```python
def Valores_Diccionario(a):
  X = a**2 + 5
  Y = a + 2
  return {"Valor_X":X, "Valor_Y":Y}
z = Valores_Diccionario (3)
print (z,type(z))
```

##Uso de una **Clase** para devolver multiples valores de una Función
Las clases contienen diferentes miembros de datos y funciones y nos permiten crear objetos para acceder a estos miembros. Podemos devolver un objeto de tales clases definidas en la función de la estructura de clases y sus miembros de datos.    

```python
class devolver_valores:
  def __init__ (self,a, b):
     self.a=a
     self.b=b
def devolver_va (a):
  b= a**2+1
  c= a + 2
  y= devolver_valores(b,c)
  return y
x=devolver_va (3)
print (x.a, x.b,type(x))
```

## Funciones anónimas o lambda:
Se define como: **lambda** parámetros: acción

Para crear una función lambda lo primero que hacemos es llamar a lambda para decir que vamos a crear una función anónima, después pasamos los parámetros que se pasarán separados por comas y por último la acción que realizará.

Ejemplo: considerando que queremos realizar una función que multiplique dos números y nos retorne un resultado, podríamos hacerlo de la siguiente forma:

```python

if __name__ == "__main__":
    
    multiplica = lambda x, y: x * y

    print(multiplica(8, 2))
```

##Ejemplo 2:

##Ejemplo 3: 
Función lambda que revise si un número es positivo

```python
lista = [18, -3, 5, 0, -1, 12, 17, 0.01, 20, -5]
lista_nueva = list(filter(lambda x: x > 0, lista))
print(lista_nueva) # [crear una lista de números positivos].
```

##Ejemplo 4: 
Función **lambda** que triplique sus argumentos de una lista

```python
j = [1,2,3,4,5,6,7,8]
nuevaLista = list(map(lambda x: x * 3, j))
print(nuevaLista)  # [imprime la lista, cada valor es multiplica por tres]
```

##Ejemplo 5: 
Función **lambda** Si necesitamos funciones que multipliquen diferentes números predeterminados, por ejemplo, duplicar, triplicar,...

```python
def multiplicar_por (n):
  return lambda x: x * n
  
duplicar = multiplicar_por(2)
triplicar = multiplicar_por(3)
diez_veces = multiplicar_por(10)

print(duplicar(10))
print(triplicar(10))
print(diez_veces(10))
```
